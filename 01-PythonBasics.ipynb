##### 物件運算子
* \_\_add\_\_   => +
* \_\_iadd\_\_  => +=
* \_\_sub\_\_ =>  -
* \_\_isub\_\_ => -=
* \_\_mul\_\_ =>  *
* \_\_imul\_\_ =>  *=
* \_\_truediv\_\_ =>  /
* \_\_itruediv\_\_  => /=
* \_\_floordiv\_\_ => //
* \_\_ifloordiv\_\_ => //=
* \_\_mod\_\_ => %
* \_\_imod\_\_ => %=
* \_\_pow\_\_ => **
* \_\_ipow\_\_ =>  **=

```python

class Person:
    def __init__(self, name, age,weight,height):  ##物件建立之後所要建立的初始化動作
        self.name = name
        self.age = age
        self.weight = weight
        self.height = height
        
    def BMI(self):
        return self.weight / ((self.height/100)**2)
    
    def __gt__(self,other):
        return float(self.BMI()) > float(other.BMI())
    
    def __add__(self,other):
        return self.height + other.height
    
    def __iadd__(self,other):
        self.height += other.height
        self.weight += other.weight
        return self

p1 = Person("John", 36,70,170)
p2 = Person("Nick", 32,80,160)
print(p1.BMI())
print(p2.BMI())
print(p1<p2)
print(p1>p2)
print(p1+p2)
print(p1.height,p1.weight)
p1+=p2
print(p1.height,p1.weight)
```

```python
from xml.dom.minidom import parse
import xml.dom.minidom

DOMTree = xml.dom.minidom.parse("data/movies.xml")
collection = DOMTree.documentElement
if collection.hasAttribute("class"):
   print ("Root element : %s" % collection.getAttribute("class"))

movies = collection.getElementsByTagName("movie")

for movie in movies:
   print ("*****Movie*****")
   if movie.hasAttribute("title"):
      print ("電影: %s" % movie.getAttribute("title"))

   type = movie.getElementsByTagName('type')[0]
   print ("類型: %s" % type.childNodes[0].data)
   format = movie.getElementsByTagName('format')[0]
   print ("格式: %s" % format.childNodes[0].data)
   starts = movie.getElementsByTagName('stars')[0]
   print ("評價: %s 顆星" % starts.childNodes[0].data)
   description = movie.getElementsByTagName('description')[0]
   print ("描述: %s" % description.childNodes[0].data)
```
