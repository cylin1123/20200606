### NLP自然語言處理

#### Jieba 中文斷字詞

* <b>規則斷詞</b> <br>
  主要是透過詞典，在對句子進行斷詞的時候，將句子的每個字與詞典中的詞進行匹配，找到則斷詞，否則無法斷詞。
* <b>統計斷詞</b><br>
  主要是看如果相連的字在不同的文本中出現的次數越多，就推斷這相連的字很可能就是一個詞。因此就可以利用字與字相鄰出现的頻率來做統計。當高於某一個臨界值時，便可認為此字組是一個詞語。

#### 精確模式 (預設模式)
將句子最精確的切開，適合文本分析

#### 全模式
把句子中所有的可以成詞的詞語都斷出来，速度快

#### 搜索引擎模式
精確模式的基礎上，對長的詞語再次切分，提高召回率，適合用於搜索引擎分詞

#### 支援自定義詞典

```python
import jieba

sentenceA = '今天星期六,我來到花蓮慈濟科技大學參加課程活動!'
sentenceB = '今天星期六,我来到花莲慈济科技大学参加课程活动!'

# 精確模式
seg_listA = jieba.cut(sentenceA) #  .lcut() return list
print("預設模式 (繁體): " + " | ".join(seg_listA))
seg_listB = jieba.cut(sentenceB) 
print("預設模式 (簡體): " + " | ".join(seg_listB))

# 全模式
seg_listA = jieba.cut(sentenceA, cut_all=True)
print("全模式: " + " | ".join(seg_listA)) 
seg_listB = jieba.cut(sentenceB, cut_all=True)
print("全模式: " + " | ".join(seg_listB)) 

# 搜尋引擎模式
seg_listA = jieba.cut_for_search(sentenceA)  
print("搜尋引擎模式 (繁體): " + "| ".join(seg_listA))
seg_listB = jieba.cut_for_search(sentenceB)  
print("搜尋引擎模式 (簡體): " + "| ".join(seg_listB))
```

```python
import jieba.posseg as pseg

sentence ='中央流行疫情指揮中心宣布，台灣今天沒有新增新冠肺炎（2019冠狀病毒疾病，COVID-19）確診，連續52天沒有本土個案，全台累計443例，428人解除隔離。'
words = pseg.cut(sentence)

for word, flag in words:
     print('%s %s' % (word, flag))
```

#### 關鍵字提取

* 基於 TF-IDF 演算法的關鍵詞抽取

    jieba.analyse.extract_tags(sentence, topK=20, withWeight=False, allowPOS=())



```python
import jieba.analyse

tags = jieba.analyse.extract_tags(sentence)

print(",".join(tags))
```

* 基於 TextRank 演算法的關鍵詞抽取

```python
tags = jieba.analyse.textrank(sentence)

keywords = ",".join(tags)
print(keywords)
```

### K-MEANS 文本分群 

安裝套件

```python
import pandas
newsdf = pandas.read_excel('data/news2.xlsx')
```

```python
newsdf.head()
```

```python
import jieba

title_ary = []
article_ary = []

for news in newsdf.iterrows():
    print(news[1])
    article_ary.append(' '.join(jieba.cut(news[1].summary)))
    title_ary.append(news[1].title)
```

```python
from sklearn.feature_extraction.text import CountVectorizer  # feature_extraction.text 文本特徵提取
vectorizer = CountVectorizer() #詞語轉換成詞頻矩陣
X = vectorizer.fit_transform(article_ary) #計算各詞語出現次數
```

```python
print(X)  # (文本序號,詞序號) 詞頻
```

```python
feature_name = vectorizer.get_feature_names()
feature_name
```

```python
X.toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_similarities = cosine_similarity(X,X)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 8, init = 'k-means++', random_state = 123)
#kmeans = KMeans(n_clusters = 8)
y_kmeans = kmeans.fit_predict(cosine_similarities)
```

```python
import numpy 
from collections import Counter
title_npary = numpy.array(title_ary)
Counter(y_kmeans)

y_kmeans
```

```python
for i in range(0,8):
    print("cluste -",i)
    print("-----------------")
    for ele in title_npary[y_kmeans == i]:    
        print(i," ",ele)
    print("-----------------")
```
